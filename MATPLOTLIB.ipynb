```python
# Pie chart 
labels = 'Frog', 'Hogs','Dogs','Logs'
sizes = [15, 30, 45, 10]
explode = (0,0.1,0,0)

figl, axl =plt.subplots()
axl.pie(sizes,explode=explode, labels=labels,autopct='%1.2f%%',
        shadow=True,
        startangle=90)
axl.axis('equal')
plt.show()
```
